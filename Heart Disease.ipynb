```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./13.1 heart-disease.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.plot.hist(figsize=(10,20),subplots=True)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
df.plot(kind="scatter",
       x="age",
       y="chol",
        c="target",
       ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=df["age"],
                  y=df["chol"],
                  c=df["target"])
ax.set(title="Age vs Chol",
      xlabel="Age",
      ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="target")
ax.axhline(df['chol'].mean(),
          linestyle="--");
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,6))
scatter=ax0.scatter(x=df["age"],
                   y=df["chol"],
                   c=df["target"])
bar=ax1.bar(df["age"],
           df["chol"])
```
